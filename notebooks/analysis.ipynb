## Text Analysis (Sentiment analysis of Headlines)

```python
#data = pd.DataFrame({'headline': ["do not buy stocks", 'being in srock market is great', 'Stocks That Hit 52-Week Highs On Friday', 'Company faces backlash over environmental concerns', "Stock markets rally as investors cheer positive..."]})

def check_polarity(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return("positive")
    elif polarity < 0:
        return('Negative')
    else:
        return('Neutral') 


ratings['sentiment'] = ratings['headline'].apply(check_polarity);

#ratings.head()
```
